***Task 1: Data Cleaning and Preprocessing***

**Description:** Work with a raw dataset that contains missing values,
duplicates, and inconsistent data formats.

**Objectives:**
- Load the dataset using pandas
- Identify and handle missing values (imputation or removal)
- Remove duplicate rows and standardize inconsistent data formats

**Tools:** Python, Pandas

```python
import pandas as pd

# Update the path to your file
file_path = '/content/drive/MyDrive/Datasets/Codveda Technology/iris.csv'

try:
    df = pd.read_csv(file_path)
    display(df.head())
except Exception as e:
    print(f"Error: {e}. Please check the file path.")
print(df.head())
print(df.shape)
print(df.info())

# Step 2: Identify missing values
print(df.isnull().sum())

# Step 3: Handle missing values
# Option A – Fill with mean (numeric columns)
df['sepal_length'].fillna(df['sepal_length'].mean(), inplace=True)
# Option B – Drop rows with missing values
df.dropna(inplace=True)

# Step 4: Remove duplicate rows
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Step 5: Standardize inconsistent formats
# Example: strip whitespace from string columns
df['species'] = df['species'].str.strip().str.lower()


# Step 6: Check data types and fix
print(df.dtypes)
# Convert if needed: df['col'] = pd.to_datetime(df['col'])


print("Cleaned dataset shape:", df.shape)
df.to_csv("iris_cleaned.csv", index=False)
```

***Task 2: Exploratory Data Analysis (EDA)***

**Description:** Perform an exploratory analysis on the cleaned dataset
to identify patterns, trends, and summary statistics.

**Objectives:**
- Calculate summary statistics (mean, median, mode, standard deviation)
- Visualize data distributions using histograms, boxplots, and scatter plots
- Find correlations between numerical features

**Tools:** Python, Pandas, Matplotlib, Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

## Task 2: Exploratory Data Analysis (EDA)

# Step 1: Summary statistics
print("=== Summary Statistics ===")
print(df.describe())
print("\nMode:\n", df.mode().iloc[0])
print("\nMedian:\n", df.median(numeric_only=True))

# Step 2: Histogram
df.hist(figsize=(10, 6), bins=20, color='steelblue')
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.show()

# Step 3: Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('species', axis=1))
plt.title("Boxplots - Outlier Detection")
plt.show()

# Step 4: Scatter plot
sns.pairplot(df, hue='species')
plt.suptitle("Pairplot", y=1.02)
plt.show()

# Step 5: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

print(" EDA Done!")
```

***Task 3: Basic Data Visualization***

**Description:** Create basic plots and charts to visualize the distribution
and relationships within the dataset.

**Objectives:**
- Create bar plots, line charts, and scatter plots
- Customize plot labels, titles, and legends
- Export plots as images for reports

**Tools:** Python, Matplotlib, Seaborn

```python
## Task 3: Basic Data Visualization

# Plot 1: Bar Plot – Species Count

species_counts = df['species'].value_counts()

plt.figure(figsize=(7, 5))
species_counts.plot(kind='bar', color=['#4C72B0', '#DD8452', '#55A868'])
plt.title("Species Count in Iris Dataset", fontsize=14, fontweight='bold')
plt.xlabel("Species", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(labels=species_counts.index, title="Species")
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Datasets/Codveda Technology/bar_plot.png', dpi=150)
plt.show()
print("Bar plot saved!")

# Plot 2: Line Chart – Feature values for first 30 rows

plt.figure(figsize=(10, 5))
df.iloc[:30][['sepal_length', 'petal_length', 'sepal_width', 'petal_width']].plot(
    figsize=(10, 5),
    marker='o'
)
plt.title("Feature Values for First 30 Rows", fontsize=14, fontweight='bold')
plt.xlabel("Index", fontsize=12)
plt.ylabel("Value (cm)", fontsize=12)
plt.legend(title="Features", loc='upper right')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Datasets/Codveda Technology/line_chart.png', dpi=150)
plt.show()
print("Line chart saved!")


# Plot 3: Scatter Plot – Sepal vs Petal Length

plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species',
                palette='Set1', s=80, edgecolor='black')
plt.title("Sepal Length vs Petal Length by Species", fontsize=14, fontweight='bold')
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.legend(title="Species", loc='upper left')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Datasets/Codveda Technology/scatter_plot.png', dpi=150)
plt.show()
print(" Scatter plot saved!")
```
