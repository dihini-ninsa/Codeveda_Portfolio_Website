```python
# Mount Drive & Import Libraries
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Datasets/Codveda Technology/Stock Prices Data Set.csv')  # Upload the file first
print("Shape:", df.shape)
df.head()
```

Data Understanding & Cleaning
Python

```python
# Basic info
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

# Convert Date to datetime
df['date'] = pd.to_datetime(df['date'])

# Check unique symbols
print("Number of unique stocks:", df['symbol'].nunique())
print("\nTop 10 Stocks by Volume:\n", df.groupby('symbol')['volume'].sum().nlargest(10))
```

```python
# Daily Return
df['Daily_Return'] = df.groupby('symbol')['close'].pct_change() * 100

# 30-day Moving Average
df['MA30'] = df.groupby('symbol')['close'].transform(lambda x: x.rolling(30).mean())

print("Feature Engineering Completed!")
df.head()
```

## Export Cleaned Dataset for Power BI
